```python
# create a function for scraping:
def scrape_hot100 (url:str, song_title_css:str, song_artist_css:str)->pd.DataFrame:
    '''
    Take 3 string inputs: url link of the website where you want to scrape, a css selector string for song_title, and a css selector string for song_artist.
    The function perform the scraping and get the texts from the css selectors, create and return a dataframe from these songs and artists
    
    input:
    url: string
    song_title_css: string
    song_artist_css: string

    output:
    df: Pandas DataFrame
    '''
    
    response = requests.get(url)
    if response.status_code == 200: # 200 status code means OK!
        soup = BeautifulSoup(response.content, "html.parser")
        song_title = [elem.get_text().strip() for elem in soup.select(song_title_css)]
        song_artist = [elem.get_text().strip() for elem in soup.select(song_artist_css)]
        hot_song_df = pd.DataFrame({"Song_title": song_title, "Artist": song_artist } )
    return hot_song_df

```

```python
# test scrape_hot100 function:
url = "https://www.billboard.com/charts/hot-100/"
song_title_css = 'li ul li h3'
song_artist_css = 'li.lrv-u-width-100p ul li h3#title-of-a-story + span'

scrape_hot100(url, song_title_css, song_artist_css)
```
